```python
import pandas as pd
import numpy as np
import re
import dill
import tqdm
import gc
import gensim
```

```python
with open('Xtrain_norm.pkl', 'rb') as f:
    X_train = []
    while True:
        try:
            X_train.extend(dill.load(f))
        except:
            print('EOF reached')
            break
            
with open('Xtest_norm.pkl', 'rb') as f:
    X_test = []
    while True:
        try:
            X_test.extend(dill.load(f))
        except:
            print('EOF reached')
            break
            
with open('ytrain_labels.pkl', 'rb') as f:
    y_train = dill.load(f)
    
with open('ytest_labels.pkl', 'rb') as f:
    y_test = dill.load(f)
    
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
import nltk
# tokenize sentences in corpus
wpt = nltk.WordPunctTokenizer()
tokenized_train = [wpt.tokenize(document) for document in X_train]
tokenized_test = [wpt.tokenize(document) for document in X_test]
```

```python
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
from gensim.models import Word2Vec

# Set values for various parameters
feature_size = 300    # Word vector dimensionality  
window_context = 50          # Context window size                                                                                    
min_word_count = 5   # Minimum word count                        
sample = 1e-3   # Downsample setting for frequent words

w2v_model = Word2Vec(tokenized_train, size=feature_size, sg=1, workers=4,
                     window=window_context, min_count=min_word_count,
                     sample=sample, iter=10)
```

```python
w2v_model.save('basic_w2v_sg10ep.bin')
```
